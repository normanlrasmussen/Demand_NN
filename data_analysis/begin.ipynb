```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load the data
df = pd.read_csv('../data/demand_data.csv')
```

```python
print(df.head())
print(df.dtypes)
```

```python
# Visualize the demand for a each product over time for each store

fig, axs = plt.subplots(10, 1, figsize=(15, 30))

for i in range(10):
    axs[i].plot(
        df[(df['store'] == i + 1) & (df['item'] == 1)]['date'],
        df[(df['store'] == i + 1) & (df['item'] == 1)]['sales']
    )
```
