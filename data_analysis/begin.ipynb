```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load the data
df = pd.read_csv('../data/demand_data.csv')
```

```python
print(df.head())
print(df.dtypes)
```

```python
# Visualize the demand for a each product over time for each store

fig, axs = plt.subplots(10, 1, figsize=(15, 30))

for i in range(10):
    axs[i].plot(
        df[(df['store'] == i + 1) & (df['item'] == 1)]['sales']
    )
```

```python
fig, axs = plt.subplots(10, 5, figsize=(15, 25))

for item_idx in range(50):
    row, col = item_idx // 5, item_idx % 5
    ax = axs[row, col]
    for store in range(1, 11):
        mask = (df['item'] == item_idx + 1) & (df['store'] == store)
        item_data = df[mask].sort_values('date')
        ax.plot(item_data['sales'])
plt.show()
```
